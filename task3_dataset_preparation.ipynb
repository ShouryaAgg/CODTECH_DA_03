```python
import pandas as pd
import numpy as np
```

```python
df_clean = pd.read_csv("clean_online_retail_2009_2011.csv")

# Convert date column back to datetime
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'])
```

```python
df_new = df_clean.copy()
```

```python
df_new = df_new.drop(columns=['InvoiceDate','Year','Month','MonthName','YearMonth','InvoiceNo'])
```

```python
# Create random dates between 2019 and 2021
start_date = pd.to_datetime("2019-01-01")
end_date = pd.to_datetime("2021-12-31")

random_dates = pd.to_datetime(
    np.random.randint(
        start_date.value // 10**9,
        end_date.value // 10**9,
        len(df_new)
    ),
    unit='s'
)

df_new['InvoiceDate'] = random_dates
```

```python
df_new = df_new.sort_values('InvoiceDate').reset_index(drop=True)
```

```python
invoice_counter = 10000
invoice_numbers = []

current_invoice = invoice_counter

for i in range(len(df_new)):
    if i == 0 or np.random.rand() < 0.15:
        current_invoice += 1
    invoice_numbers.append(current_invoice)

df_new['InvoiceNo'] = invoice_numbers
```

```python
df_new['Year'] = df_new['InvoiceDate'].dt.year
df_new['Month'] = df_new['InvoiceDate'].dt.month
df_new['MonthName'] = df_new['InvoiceDate'].dt.month_name()
df_new['YearMonth'] = df_new['InvoiceDate'].dt.to_period('M')
```

```python
print(df_new['Year'].value_counts())
print(df_new['InvoiceDate'].min())
print(df_new['InvoiceDate'].max())
```

```python
df_new.to_csv("online_retail_2019_2021.csv", index=False)
```
